# Gradient Descent

### 라이브러리 import 및 설정

```python
import csv
import random
import pylab as plt
```

### csv 파일에서 필요한 값들 가져오기

```python
x = []
y = []

f = open('hw1_data.csv', 'r', encoding='utf-8')
rdr = csv.reader(f)

i = 0
for line in rdr:
    i = i+1
    if i == 1:
        continue
    x.append(float(line[3]))
    y.append(float(line[7]))
f.close()
```

### 데이터 분포 확인

```python
m = len(list(y))

print("max(x):", max(x), "min(x):", min(x))
print("max(y):", max(y), "min(y):", min(y))
```

```python
plt.plot(x,y,'o')
plt.xlim([-1000,6500]); plt.ylim([0,117])
plt.axvline(x=0, c='g'); plt.axhline(y=0, c='g')
plt.show()
```

### 세타, 알파 초기값 지정

```python
theta0 = random.randint(40,50)
theta3 = random.random()
alpha = 0.0000001

print("theta0:", theta0)
print("theta3:", theta3)
```

### 경사하강법에 사용되는 함수들

```python
def h(a,b,x):
    return a*x + b

def sum(a,b,x,y):
    s = 0
    for i,j in zip(x,y):
        s = s + ((h(a,b,i) - j) ** 2)
    return s

def sum_theta0(a,b,x,y):
    s = 0
    for i,j in zip(x,y):
        s = s + (h(a,b,i) - j)
    return s
        
def sum_theta3(a,b,x,y):
    s = 0
    for i,j in zip(x,y):
        s = s + ((h(a,b,i) - j) * i)
    return s

def j(x,y):
    return (1 / (2*m)) * sum(theta3,theta0,x,y)
```

### Cost Function 최소화

```python
iteration = 0

while 1:
    iteration = iteration + 1
    
    print("#",iteration," theta3: ",theta3, "theta0: ",theta0)
    print("       j(x,y): ",j(x,y),"\n")
    
    origin = j(x,y)
    
    updated_theta0 = theta0 - alpha * (1/m) * sum_theta0(theta3,theta0,x,y)
    updated_theta3 = theta3 - alpha * (1/m) * sum_theta3(theta3,theta0,x,y)
    
    theta0 = updated_theta0
    theta3 = updated_theta3
    
    if origin < j(x,y):
        print("Set the alpha again.")
        break
    
    if iteration > 10000:
        break
```

### 최종 방정식 그래프 확인하기

```python
f_x = [f_a for f_a in range(-6000,6000)]
f_y = [theta3*f_b + theta0 for f_b in range(-6000,6000)]

plt.plot(f_x,f_y)
plt.plot(x,y,'o')
plt.xlim([-1000,6000]); plt.ylim([0,100])
plt.axvline(x=0, c='g'); plt.axhline(y=0, c='g')
plt.show()
```
