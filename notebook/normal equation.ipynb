# Normal Equation

### 라이브러리 import 및 설정

```python
import csv
import numpy
import numpy as np
```

### csv 파일에서 필요한 값들 가져오기

```python
X_list = [[], [], []]
y_list = []

f = open('hw1_data.csv', 'r', encoding='utf-8')
rdr = csv.reader(f)

i = 0
for line in rdr:
    i = i+1
    if i == 1:
        continue
    X_list[0].append(1)
    X_list[1].append(float(line[2]))
    X_list[2].append(float(line[3]))
    y_list.append(float(line[7]))
    
X,y =np.array(X_list).transpose(),np.array(y_list)

f.close()
```

### 행렬 출력

```python
print("X: \n", X, "\n")
print("y: \n", y)
```

### 정규화

```python
theta = numpy.dot(numpy.dot(numpy.linalg.inv(numpy.dot(X.transpose(),X)), X.transpose()), y)
```

### 세타값 출력

```python
print("theta0:", theta[0], " theta2:",theta[1], " theta3:",theta[2])
```
